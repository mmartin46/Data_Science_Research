```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Mitchell_Martin_Budget_2024_ML.csv')
```

```python
def get_filtered_dataset(df):
  df.dropna(axis='columns', inplace=True)
  # Drop the first row
  df = df.drop(df.index[0])

  # Convert date to a date format
  df['Date'] = pd.to_datetime(df['Date'])
  df['Remainder'] = 0
  remove_dollar_signs(df)
  return df

# Makes the columns numerical
def remove_dollar_signs(df):
  numeric_cols = list(df.columns)[3:-1]
  for col in numeric_cols:
    df[col] = pd.to_numeric(df[col].str.replace('[\$,]', '', regex=True), errors='coerce')


df = get_filtered_dataset(df)
df
```
